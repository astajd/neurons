```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

from utils import *
```

```python
# load image
im = Image.open('testim.png').convert('L')

# assuming 0.452um/px resolution, rescale the image if needed or use  
# anisodiff function with different parameters (see equation 6 in chapter 2.2)

#scaling = 2
#im = im.resize((int(im.size[0]/scaling), int(im.size[1]/scaling)), Image.ANTIALIAS)
im = np.array(im)
```

```python
f = plt.figure(figsize=(7,7))
plt.imshow(im, cmap='gray')
plt.show()
```

```python
# apply the anisotropic diffusion with default parameters
im_aniso = anisodiff(im)
```

```python
# obtain locations from diffused image 
pts = getPts(im_aniso)
```

```python
f = plt.figure(figsize=(7,7))
plt.imshow(im, cmap='gray')

plt.plot( [x[1] for x in pts], [x[0] for x in pts], 'w.', markersize=5)

plt.tight_layout()
plt.show()
```
